```python
import pandas as pd
```

```python
def remove_duplicate_lines(input_file, output_file):
    with open(input_file, 'r') as file:
        lines = file.readlines()

    unique_lines = []
    for i in range(len(lines)):
        if i % 2 == 0: 
            unique_lines.append(lines[i].strip())

    with open(output_file, 'w') as file:
        for line in unique_lines:
            file.write(f"{line}\n")

input_filename = '../testResults/theta_star_expandedVertices.txt'  
output_filename = '../testResults/theta_star_expanded.txt'  
remove_duplicate_lines(input_filename, output_filename)
```

```python
input_file = '../testResults/theta_star_expanded.txt'

df1 = pd.read_csv(input_file, header=None, names=['Expanded'])
df1['Index'] = df1.index
df1 = df1[['Index', 'Expanded']]

input_file = '../testResults/theta_star_time_length.txt'

df2 = pd.read_csv(input_file, header=None, names=['Time', 'Length'], sep=' ')
df2['Index'] = df2.index
df2 = df2[['Index', 'Time', 'Length']]

merged_df = pd.merge(df1, df2, on='Index', how='inner')
output_file = 'theta_star_metrics.csv'
merged_df.to_csv(output_file, index=False)

merged_df
```

```python
input_file = '../testResults/a_star_expanded.txt'

df3 = pd.read_csv(input_file, header=None, names=['Expanded'])
df3['Index'] = df3.index
df3 = df3[['Index', 'Expanded']]

input_file = '../testResults/a_star_time_length.txt'

df4 = pd.read_csv(input_file, header=None, names=['Time', 'Length'], sep=' ')
df4['Index'] = df4.index
df4 = df4[['Index', 'Time', 'Length']]

merged_df2 = pd.merge(df3, df4, on='Index', how='inner')
output_file = 'a_star_metrics.csv'
merged_df2.to_csv(output_file, index=False)

merged_df2
```

```python
maps = ["AR0500SR", "maze512-2-5", "Milan_1_1024", "random512-20-0"]

input_file = 'theta_star_metrics.csv'
df = pd.read_csv(input_file, header=None, names=['Index', 'Expanded', 'Time', 'Length'], sep=',')
df = df.drop(index=0).reset_index(drop=True)
print(df)

tasks_per_map = 200

for i in range(4):
    start_index = i * tasks_per_map
    end_index = start_index + tasks_per_map
    chunk = df.iloc[start_index : end_index]

    file_name = f"theta_star_{maps[i]}_result.csv"  
    chunk.to_csv(file_name, index=False)
```

```python
input_file = 'a_star_metrics.csv'
df = pd.read_csv(input_file, header=None, names=['Index', 'Expanded', 'Time', 'Length'], sep=',')
df = df.drop(index=0).reset_index(drop=True)
print(df)

for i in range(4):
    start_index = i * tasks_per_map
    end_index = start_index + tasks_per_map
    chunk = df.iloc[start_index : end_index]
    
    file_name = f"a_star_{maps[i]}_result.csv"  
    chunk.to_csv(file_name, index=False)
```
